```python
import pandas as pd
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
sns.set(style='whitegrid')
```

```python
# Define the file path
csv_path = '../data/menu.csv'

# Read the CSV file into a DataFrame
data = pd.read_csv(csv_path)

# Display the first 5 rows
data.head()
```

```python
# Connect to SQLite database (it will be created if it doesn't exist)
conn = sqlite3.connect('../db/mcdonalds.db')

# Store the DataFrame as a SQL table
data.to_sql('mcdonalds_nutrition', conn, if_exists='replace', index=False)

# Confirm the table is stored
print("Table 'mcdonalds_nutrition' has been saved to the database.")
```

```python
query = "SELECT * FROM mcdonalds_nutrition"
df = pd.read_sql(query, conn)
df.head()
```

```python
print("Dataset Shape:", df.shape)
df.info()
df.describe()
```

```python
max_sodium_idx = df['Sodium'].idxmax()
df.loc[[max_sodium_idx], ['Item', 'Sodium']]
```

```python
plt.figure(figsize=(14, 6))
sns.swarmplot(x='Category', y='Sodium', data=df)
plt.xticks(rotation=45)
plt.title('Sodium Content by Food Category')
plt.tight_layout()
plt.show()
```

```python
sns.jointplot(x='Protein', y='Total Fat', data=df, kind='scatter')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(y='Sugars', data=df)
plt.title('Sugar Content Distribution')
plt.show()
```

### Summary

- The dataset includes nutritional facts for McDonald's menu items.
- Stored the data into an SQLite database using Python.
- Found that **Chicken McNuggets (40 pieces)** has the highest sodium content (3600 mg).
- Protein and Total Fat are **strongly positively correlated**.
- Sugar content has **high outliers**, likely from desserts or sweet drinks.
